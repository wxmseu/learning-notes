pandas数据排序

series排序：
series.sort_values(ascending=True, inplace=Fals)
参数说明：
    ascending：默认为true升序排序，为false降序排序
    inplace：是否修改原始series
df的排序
df.sort_values(by,ascending=True,inplace=True)
参数说明：
    by：字符串或者list，单列排序或者多列排序
    ascending：bool或者list，升序还是降序，如果是list对应by的多列
    inplace：是否修改原始df


```python
import pandas as pd
df=pd.read_excel('weather.xlsx')
```

1. series的排序

```python
df['最低温'].sort_values()
```

```python
df['最低温'].sort_values(ascending=False)
```

2. df的排序

2.1 单列排序

```python
df.sort_values(by='最低温')
```

2.2 多列排序

```python
# 按最高温升序，最低温降序排序
df.sort_values(by=['最高温','最低温'],ascending=[True,False])
```

```python
# 两个字段都按照降序排序
df.sort_values(by=['最高温','最低温'],ascending=False)
```
